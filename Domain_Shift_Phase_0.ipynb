## Visualization

```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# ============================================
# 2D SLICES VISUALIZATION
# ============================================

def visualize_2d_slices(image_path, label_path=None, num_slices=8, axis=2,
                        cmap_image='gray', cmap_label='jet', title="2D Slices"):
    """
    Visualize 2D slices from 3D volume with optional segmentation overlay.

    Args:
        image_path: Path to image NIfTI file
        label_path: Optional path to label NIfTI file
        num_slices: Number of slices to display
        axis: Axis along which to slice (0=sagittal, 1=coronal, 2=axial)
        cmap_image: Colormap for image
        cmap_label: Colormap for labels
        title: Plot title
    """
    # Load data
    img_data = nib.load(image_path).get_fdata()

    if label_path and Path(label_path).exists():
        lbl_data = nib.load(label_path).get_fdata()
        has_labels = True
    else:
        lbl_data = None
        has_labels = False

    # Calculate slice indices
    slice_dim = img_data.shape[axis]
    step = max(1, slice_dim // (num_slices + 1))
    slice_indices = [step * (i + 1) for i in range(num_slices)]

    # Create figure
    rows = 2 if has_labels else 1
    fig, axes = plt.subplots(rows, num_slices, figsize=(18, 4*rows))

    if rows == 1:
        axes = axes.reshape(1, -1)

    for i, idx in enumerate(slice_indices):
        # Extract slice based on axis
        if axis == 0:
            img_slice = img_data[idx, :, :]
            lbl_slice = lbl_data[idx, :, :] if has_labels else None
        elif axis == 1:
            img_slice = img_data[:, idx, :]
            lbl_slice = lbl_data[:, idx, :] if has_labels else None
        else:  # axis == 2
            img_slice = img_data[:, :, idx]
            lbl_slice = lbl_data[:, :, idx] if has_labels else None

        # Plot image
        axes[0, i].imshow(img_slice.T, cmap=cmap_image, origin='lower')
        axes[0, i].set_title(f'Slice {idx}', fontsize=10)
        axes[0, i].axis('off')

        # Plot label if available
        if has_labels:
            # Overlay: show image with label overlay
            axes[1, i].imshow(img_slice.T, cmap=cmap_image, origin='lower', alpha=0.7)
            if np.any(lbl_slice):
                axes[1, i].imshow(np.ma.masked_where(lbl_slice.T == 0, lbl_slice.T),
                                 cmap=cmap_label, origin='lower', alpha=0.5)
            axes[1, i].set_title(f'Overlay {idx}', fontsize=10)
            axes[1, i].axis('off')

    fig.suptitle(title, fontsize=14, fontweight='bold', y=0.98)
    plt.tight_layout()
    plt.show()


# ============================================
# USAGE EXAMPLE
# ============================================

# Define paths
source_images_path = Path("/content/Dataset996_MRA_Aneurysm/imagesTr")
source_labels_path = Path("/content/Dataset996_MRA_Aneurysm/labelsTr")

target_images_path = Path("/content/Dataset511_TopcowMR/imagesTr")
target_labels_path = Path("/content/Dataset511_TopcowMR/labelsTr")

# Get first file from each dataset
if source_images_path.exists() and target_images_path.exists():
    source_img_files = sorted(list(source_images_path.glob("*.nii*")))
    target_img_files = sorted(list(target_images_path.glob("*.nii*")))

    if source_img_files and target_img_files:
        source_img = source_img_files[0]
        source_lbl_name = source_img.name.replace("_0000", "")
        source_lbl = source_labels_path / source_lbl_name

        target_img = target_img_files[0]
        # Handle potential _0000 suffix in target images
        target_lbl_name = target_img.name.replace("_0000", "")
        target_lbl = target_labels_path / target_lbl_name

        print("=" * 80)
        print("VISUALIZING SOURCE DATASET (CTA) - 2D SLICES")
        print("=" * 80)
        print(f"Image: {source_img.name}")
        print(f"Label: {source_lbl.name if source_lbl.exists() else 'Not found'}\n")

        visualize_2d_slices(source_img, source_lbl, num_slices=6,
                           title="Source (RSNA) - Axial Slices")

        print("\n" + "=" * 80)
        print("VISUALIZING TARGET DATASET (TopCoW) - 2D SLICES")
        print("=" * 80)
        print(f"Image: {target_img.name}")
        print(f"Label: {target_lbl.name if target_lbl.exists() else 'Not found'}\n")

        visualize_2d_slices(target_img, target_lbl, num_slices=6,
                           title="Target (TopCoW) - Axial Slices")

        print("\n✅ 2D Visualization complete!")
    else:
        print("⚠️ No image files found in one or both directories")
else:
    print("⚠️ One or both dataset directories do not exist")
```

```python
import nibabel as nib
import numpy as np
from pathlib import Path
from nilearn import plotting

# ============================================
# INTERACTIVE 3D VISUALIZATION WITH NILEARN
# ============================================

def visualize_interactive(image_path, label_path=None, title="Interactive Viewer"):
    """
    Open an interactive 3D viewer in your browser with optional segmentation overlay.

    Args:
        image_path: Path to image NIfTI file
        label_path: Optional path to label/segmentation NIfTI file
        title: Title for the viewer

    Returns:
        Interactive viewer object (automatically opens in browser)
    """
    print(f"Opening interactive viewer: {title}")
    print(f"Image: {Path(image_path).name}")
    if label_path and Path(label_path).exists():
        print(f"Label: {Path(label_path).name}")
    print("-" * 60)

    # Load the main image
    img = nib.load(image_path)

    # If we have a label, create an overlay view
    if label_path and Path(label_path).exists():
        # Use view_img with background and overlay
        viewer = plotting.view_img(
            stat_map_img=label_path,  # The segmentation as overlay
            bg_img=image_path,         # The original image as background
            title=title,
            cmap='jet',                # Colormap for the segmentation
            opacity=0.5,               # Transparency of overlay
            resampling_interpolation='nearest'  # Keep label values intact
        )
    else:
        # Just show the image without overlay
        viewer = plotting.view_img(
            stat_map_img=image_path,
            title=title,
            cmap='gray'
        )

    return viewer


# ============================================
# USAGE EXAMPLE WITH YOUR DATA
# ============================================

# Define paths
source_images_path = Path("/content/Dataset996_MRA_Aneurysm/imagesTr")
source_labels_path = Path("/content/Dataset996_MRA_Aneurysm/labelsTr")

target_images_path = Path("/content/Dataset511_TopcowMR/imagesTr")
target_labels_path = Path("/content/Dataset511_TopcowMR/labelsTr")

# Get first file from each dataset
if source_images_path.exists() and target_images_path.exists():
    source_img_files = sorted(list(source_images_path.glob("*.nii*")))
    target_img_files = sorted(list(target_images_path.glob("*.nii*")))

    if source_img_files and target_img_files:
        source_img = source_img_files[0]
        source_lbl_name = source_img.name.replace("_0000", "")
        source_lbl = source_labels_path / source_lbl_name

        target_img = target_img_files[0]
        # Handle potential _0000 suffix in target images
        target_lbl_name = target_img.name.replace("_0000", "")
        target_lbl = target_labels_path / target_lbl_name

        print("=" * 80)
        print("INTERACTIVE VISUALIZATION - SOURCE DATASET (CTA)")
        print("=" * 80)

        # Open interactive viewer for SOURCE dataset
        source_viewer = visualize_interactive(
            source_img,
            source_lbl if source_lbl.exists() else None,
            title="Source (CTA) - Interactive 3D Viewer"
        )

        # Display in Colab notebook
        from IPython.display import display
        display(source_viewer)

        print("\n" + "=" * 80)
        print("INTERACTIVE VISUALIZATION - TARGET DATASET (TopCoW)")
        print("=" * 80)

        # Open interactive viewer for TARGET dataset
        target_viewer = visualize_interactive(
            target_img,
            target_lbl if target_lbl.exists() else None,
            title="Target (TopCoW) - Interactive 3D Viewer"
        )

        # Display in Colab notebook
        display(target_viewer)

        print("\n✅ Interactive viewers opened!")
        print("📌 Use your mouse to click and scroll through slices")
        print("📌 Toggle between axial, sagittal, and coronal views")

    else:
        print("⚠️ No image files found in one or both directories")
else:
    print("⚠️ One or both dataset directories do not exist")
```

## Dataset Charactrization

```python
import os
from pathlib import Path
import nibabel as nib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm

# ============================================
# METRIC CALCULATION FUNCTIONS
# ============================================

def calculate_intensity_metrics(data: np.ndarray) -> tuple:
    """
    Calculate mean and standard deviation of intensity values on foreground voxels.

    Args:
        data: 3D numpy array of image data

    Returns:
        Tuple of (mean_intensity, std_intensity)
    """
    foreground_voxels = data[(data > 0) & (data < 80)]
    if foreground_voxels.size > 0:
        mean_intensity = foreground_voxels.mean()
        std_intensity = foreground_voxels.std()
    else:
        mean_intensity = 0.0
        std_intensity = 0.0

    return mean_intensity, std_intensity


def extract_resolution(header) -> tuple:
    """
    Extract voxel spacing (resolution) from NIfTI header.

    Args:
        header: NIfTI image header

    Returns:
        Tuple of (resolution_x, resolution_y, resolution_z) in mm
    """
    resolution = header.get_zooms()[:3]
    return tuple(resolution)


def estimate_noise(data: np.ndarray) -> float:
    """
    Estimate noise level by calculating standard deviation of background voxels.
    Background is defined as voxels in the bottom 5th percentile of intensity.

    Args:
        data: 3D numpy array of image data

    Returns:
        Standard deviation of background (noise estimate)
    """
    background_threshold = np.percentile(data, 5)
    background_voxels = data[data < background_threshold]

    if background_voxels.size > 0:
        noise_std = background_voxels.std()
    else:
        noise_std = 0.0

    return noise_std


def create_intensity_histogram_data(data: np.ndarray, sample_size: int = 200000) -> np.ndarray:
    """
    Sample foreground voxel intensities for histogram plotting.

    Args:
        data: 3D numpy array of image data
        sample_size: Maximum number of samples to return

    Returns:
        Array of sampled intensity values
    """
    foreground_voxels = data[(data > 0) & (data < 80)]

    if foreground_voxels.size > 0:
        actual_sample_size = min(len(foreground_voxels), sample_size)
        samples = np.random.choice(foreground_voxels, size=actual_sample_size, replace=False)
    else:
        samples = np.array([])

    return samples


# ============================================
# COMPREHENSIVE DATASET ANALYSIS
# ============================================

def analyze_dataset(images_dir: Path, labels_dir: Path = None):
    """
    Performs comprehensive dataset characterization including:
    - Basic metrics: mean/std intensity, resolution
    - Noise estimation
    - Histogram data for visualization

    Args:
        images_dir: Path to directory containing image files (.nii or .nii.gz)
        labels_dir: Path to directory containing label files (optional, not used currently)

    Returns:
        Tuple of (DataFrame with all metrics, array of intensity samples for histogram)
    """
    all_metrics = []
    all_intensities = []

    # Search for both .nii and .nii.gz files
    image_files = sorted(list(images_dir.glob("*.nii.gz")) + list(images_dir.glob("*.nii")))

    if not image_files:
        print(f"Warning: No '.nii' or '.nii.gz' files found in {images_dir}")
        return pd.DataFrame(), np.array([])

    for img_path in tqdm(image_files, desc=f"Analyzing {images_dir.parent.name}"):
        try:
            # Load image
            img = nib.load(img_path)
            data = img.get_fdata()
            header = img.header

            # Calculate metrics using individual functions
            mean_intensity, std_intensity = calculate_intensity_metrics(data)
            resolution_x, resolution_y, resolution_z = extract_resolution(header)
            noise_std = estimate_noise(data)
            intensity_samples = create_intensity_histogram_data(data)

            # Collect samples for histogram
            all_intensities.extend(intensity_samples)

            # Store all metrics
            all_metrics.append({
                "filename": img_path.name,
                "mean_intensity": mean_intensity,
                "std_intensity": std_intensity,
                "resolution_x": resolution_x,
                "resolution_y": resolution_y,
                "resolution_z": resolution_z,
                "noise_std": noise_std
            })

        except Exception as e:
            print(f"Could not process {img_path.name}: {e}")

    return pd.DataFrame(all_metrics), np.array(all_intensities)


# ============================================
# MAIN ANALYSIS EXECUTION
# ============================================

# Define Paths
source_images_path = Path("/content/Dataset996_MRA_Aneurysm/imagesTr")
source_labels_path = Path("/content/Dataset996_MRA_Aneurysm/labelsTr")

target_images_path = Path("/content/Dataset511_TopcowMR/imagesTr")
target_labels_path = Path("/content/Dataset511_TopcowMR/labelsTr")

# Run Analysis
print("Starting dataset characterization analysis...\n")

source_df, source_intensities = analyze_dataset(source_images_path, source_labels_path)
target_df, target_intensities = analyze_dataset(target_images_path, target_labels_path)

# Display Summary Table
source_summary = source_df.mean(numeric_only=True).rename("Source (RSNA)")
target_summary = target_df.mean(numeric_only=True).rename("Target (TopCoW)")

summary_table = pd.concat([source_summary, target_summary], axis=1)
print("\n" + "="*80)
print("DATASET CHARACTERIZATION SUMMARY")
print("="*80)
print(summary_table.round(4))
print("="*80)


# --- Create Visualization Plots ---
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(1, 1, figsize=(12, 6))

# Calculate reasonable x-axis limit (99th percentile of intensities)
max_intensity = np.percentile(np.concatenate([source_intensities, target_intensities]), 99)

ax.hist(source_intensities, bins=256, density=True, alpha=0.7, label='Source (RSNA)', color='blue', range=(0, max_intensity))
ax.hist(target_intensities, bins=256, density=True, alpha=0.7, label='Target (TopCoW)', color='orange', range=(0, max_intensity))
ax.set_title('Foreground Voxel Intensity Distribution', fontsize=14, fontweight='bold')
ax.set_xlabel('Intensity Value', fontsize=12)
ax.set_ylabel('Density', fontsize=12)
ax.set_xlim(0, max_intensity)  # Limit x-axis to 99th percentile
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n✅ Analysis complete!")
```

```python
import numpy as np
import nibabel as nib
from scipy.ndimage import distance_transform_edt
from skimage.morphology import skeletonize
from pathlib import Path
import matplotlib.pyplot as plt
from tqdm import tqdm
import gc

# ============================================
# MEMORY-EFFICIENT VESSEL DIAMETER CALCULATION
# ============================================

def calculate_vessel_diameters(mask_data, voxel_spacing):
    """
    Memory-efficient vessel diameter calculation using float32.

    Args:
        mask_data: 3D NumPy array (binary segmentation mask)
        voxel_spacing: Tuple of 3 floats (x, y, z) in mm

    Returns:
        1D NumPy array of vessel diameters in mm
    """
    # Convert to uint8 to save memory
    mask_binary = (mask_data > 0).astype(np.uint8)

    # Clear original mask from memory
    del mask_data
    gc.collect()

    # Step 1: Distance Transform with float32 (HALVES MEMORY USAGE)
    radius_map = distance_transform_edt(mask_binary, sampling=voxel_spacing).astype(np.float32)

    # Step 2: Skeletonization
    skeleton = skeletonize(mask_binary.astype(bool))

    # Clear binary mask from memory
    del mask_binary
    gc.collect()

    # Step 3: Extract diameters at skeleton points
    radii_mm = radius_map[skeleton]
    diameters_mm = radii_mm * 2

    # Clear intermediate arrays
    del radius_map, skeleton, radii_mm
    gc.collect()

    return diameters_mm


# ============================================
# CHUNKED BATCH ANALYSIS - NO SAMPLING
# ============================================

def analyze_dataset_diameters(labels_dir, chunk_size=5):
    """
    Memory-efficient batch analysis with chunked processing.
    Processes ALL diameter points (no sampling).

    Args:
        labels_dir: Path to directory containing label files
        chunk_size: Number of files to process before cleanup

    Returns:
        Dictionary of results and array of ALL diameters
    """
    labels_path = Path(labels_dir)
    label_files = sorted(list(labels_path.glob("*.nii.gz")) + list(labels_path.glob("*.nii")))

    all_results = {}
    diameter_list = []

    print(f"Found {len(label_files)} label files to analyze\n")

    # Process in chunks
    num_chunks = (len(label_files) + chunk_size - 1) // chunk_size

    for chunk_idx in range(num_chunks):
        start_idx = chunk_idx * chunk_size
        end_idx = min(start_idx + chunk_size, len(label_files))
        chunk_files = label_files[start_idx:end_idx]

        print(f"\n📦 Processing chunk {chunk_idx + 1}/{num_chunks} ({len(chunk_files)} files)")

        for label_file in tqdm(chunk_files, desc=f"Chunk {chunk_idx + 1}"):
            try:
                # Load image (mmap=False prevents memory buildup)
                label_img = nib.load(label_file, mmap=False)
                mask_data = label_img.get_fdata()
                voxel_spacing = label_img.header.get_zooms()[:3]

                del label_img
                gc.collect()

                # Calculate ALL diameters (no sampling)
                diameters = calculate_vessel_diameters(mask_data, voxel_spacing)

                # Store summary statistics
                all_results[label_file.name] = {
                    'mean': float(diameters.mean()),
                    'median': float(np.median(diameters)),
                    'count': len(diameters)
                }

                # Store ALL diameters (no sampling)
                diameter_list.append(diameters)

                print(f"✅ {label_file.name}: {len(diameters)} points, mean={diameters.mean():.3f} mm")

                del diameters, mask_data
                gc.collect()

            except Exception as e:
                print(f"❌ Error processing {label_file.name}: {e}")
                gc.collect()

        # Force cleanup after each chunk
        print(f"✅ Chunk {chunk_idx + 1} complete. Running garbage collection...")
        gc.collect()

    # Combine ALL diameters
    all_diameters = np.concatenate(diameter_list) if diameter_list else np.array([])
    del diameter_list
    gc.collect()

    return all_results, all_diameters


# ============================================
# VISUALIZATION FUNCTION
# ============================================

def visualize_diameter_comparison(source_diameters, target_diameters):
    """Create comparison visualizations with dynamic limits."""
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    # Calculate dynamic limit (99th percentile)
    combined_diameters = np.concatenate([source_diameters, target_diameters])
    xlim_max = np.percentile(combined_diameters, 99)

    # Plot 1: Histogram
    axes[0].hist(source_diameters, bins=100, alpha=0.6, label='Source (RSNA)',
                 color='blue', density=True, range=(0, xlim_max))
    axes[0].hist(target_diameters, bins=100, alpha=0.6, label='Target (TopCoW)',
                 color='orange', density=True, range=(0, xlim_max))
    axes[0].set_xlabel('Diameter (mm)', fontsize=12)
    axes[0].set_ylabel('Density', fontsize=12)
    axes[0].set_title('Vessel Diameter Distribution Comparison', fontsize=14, fontweight='bold')
    axes[0].set_xlim(0, xlim_max)
    axes[0].legend(fontsize=11)
    axes[0].grid(True, alpha=0.3)

    # Plot 2: Box plot (full range, no filtering)
    box_data = [source_diameters, target_diameters]
    bp = axes[1].boxplot(box_data, labels=['Source (RSNA)', 'Target (TopCoW)'],
                         patch_artist=True, showfliers=True)
    bp['boxes'][0].set_facecolor('blue')
    bp['boxes'][1].set_facecolor('orange')
    axes[1].set_ylabel('Diameter (mm)', fontsize=12)
    axes[1].set_title('Vessel Diameter Box Plot (Full Range)', fontsize=14, fontweight='bold')
    axes[1].grid(True, alpha=0.3, axis='y')

    plt.tight_layout()
    plt.show()

    # Print statistics
    print(f"\n{'='*80}")
    print(f"VESSEL DIAMETER STATISTICS COMPARISON")
    print(f"{'='*80}")
    print(f"{'Metric':<25} {'Source (RSNA)':<20} {'Target (TopCoW)':<20}")
    print(f"{'-'*80}")
    print(f"{'Total skeleton points':<25} {len(source_diameters):<20} {len(target_diameters):<20}")
    print(f"{'Mean diameter (mm)':<25} {source_diameters.mean():<20.3f} {target_diameters.mean():<20.3f}")
    print(f"{'Median diameter (mm)':<25} {np.median(source_diameters):<20.3f} {np.median(target_diameters):<20.3f}")
    print(f"{'Std deviation (mm)':<25} {source_diameters.std():<20.3f} {target_diameters.std():<20.3f}")
    print(f"{'Min diameter (mm)':<25} {source_diameters.min():<20.3f} {target_diameters.min():<20.3f}")
    print(f"{'Max diameter (mm)':<25} {source_diameters.max():<20.3f} {target_diameters.max():<20.3f}")
    print(f"{'25th percentile (mm)':<25} {np.percentile(source_diameters, 25):<20.3f} {np.percentile(target_diameters, 25):<20.3f}")
    print(f"{'75th percentile (mm)':<25} {np.percentile(source_diameters, 75):<20.3f} {np.percentile(target_diameters, 75):<20.3f}")
    print(f"{'99th percentile (mm)':<25} {np.percentile(source_diameters, 99):<20.3f} {np.percentile(target_diameters, 99):<20.3f}")
    print(f"{'='*80}\n")


# ============================================
# RUN ANALYSIS WITH CHUNKED PROCESSING
# ============================================

# Define paths
source_labels_path = Path("/content/Dataset996_MRA_Aneurysm/labelsTr")
target_labels_path = Path("/content/Dataset511_TopcowMR/labelsTr")

print("="*80)
print("ANALYZING SOURCE DATASET (RSNA)")
print("="*80)
source_results, source_all_diameters = analyze_dataset_diameters(
    source_labels_path,
    chunk_size=5  # Only chunk size parameter
)

# Force garbage collection between datasets
gc.collect()

print("\n" + "="*80)
print("ANALYZING TARGET DATASET (TopCoW)")
print("="*80)
target_results, target_all_diameters = analyze_dataset_diameters(
    target_labels_path,
    chunk_size=5
)

# Visualize comparison
print("\n" + "="*80)
print("VISUALIZING COMPARISON")
print("="*80)
visualize_diameter_comparison(source_all_diameters, target_all_diameters)

print("\n✅ Vessel diameter analysis complete!")
```
